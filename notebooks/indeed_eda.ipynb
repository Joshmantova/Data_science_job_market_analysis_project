```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
df_all = pd.read_csv('../Datasets/df_all_indeed.csv', index_col=0)
```

```python
def get_state_co(x):
    if 'CO' in x or 'Colorado' in x or 'Denver' in x:
        return 'CO'
    else:
        return 'Not CO'
```

```python
df_all['Colorado_state'] = df_all['Locations'].apply(get_state_co)
df_all
```

```python
co_rating_mean = df_all.groupby('Colorado_state').mean()['Rating']
co_state = co_rating_mean.index
co_rating = co_rating_mean.values

fig, ax = plt.subplots()
ax.bar(co_state, co_rating)
ax.set_title('Avg Company Ratings for CO Jobs vs Non-CO Jobs')
ax.set_xlabel('State')
ax.set_ylabel('Avg Company Rating')
plt.tight_layout();
```

```python
easy_apply_rating_mean = df_all.groupby('Easy_Apply').mean()['Rating']
easy_apply = easy_apply_rating_mean.index
easy_apply_ratings = easy_apply_rating_mean.values

fig, ax = plt.subplots()
ax.bar(easy_apply, easy_apply_ratings)
ax.set_xlabel('Easy Apply or Not')
ax.set_ylabel('Avg Company Rating')
ax.set_title('Average Company Rating for Easy Apply Jobs vs Not')
plt.tight_layout();
```

```python
def get_state(x):
    if 'CO' in x or 'Colorado' in x or 'Denver' in x:
        return 'CO'
    elif 'CA' in x or 'California' in x or 'San Francisco' in x or 'Los Angeles' in x or 'Sunnyvale' in x:
        return 'CA'
    elif 'FL' in x or 'Florida' in x or 'Miami' in x:
        return 'FL'
    elif 'NY' in x or 'New York' in x:
        return 'NY'
    elif 'UT' in x or 'Utah' in x or 'Salt Lake City' in x:
        return 'UT'
    elif 'WA' in x or 'Washington' in x or 'Seattle' in x:
        return 'WA'
```

```python
df_all['State'] = df_all['Locations'].apply(get_state)
```

```python
df_all
```

```python
state_rating_means = df_all.groupby('State').mean()['Rating']
states = state_rating_means.index
ratings = state_rating_means.values
```

```python
fig, ax = plt.subplots()
ax.bar(states, ratings)
ax.set_title('Avg Company Ratings for Each State')
ax.set_xlabel('State')
ax.set_ylabel('Avg Company Rating')
plt.tight_layout();
```
