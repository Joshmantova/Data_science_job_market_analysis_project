```python
def includes_salary(descrip):
    if '$' in descrip or 'salary' in descrip:
        return 1
    else:
        return 0
    
```

```python
df['Salary_included'] = df['Description'].apply(includes_salary)
df.columns
```

```python
def get_num_applicants_int(x):
    num = str()
    for elem in x:
        if elem == 'B':
            num = 20
            break
        elif elem == 'O':
            num = 200
            break
        elif elem != ' ':
            num += elem
        elif elem == ' ':
            break

    return int(num)
            
df['num_applicants'] = df['Number_of_Applicants'].apply(get_num_applicants_int)

df['num_applicants']
```

```python
df[df['Number_of_Applicants'] != 'Be among the first 25 applicants'].count()
```

```python
def get_length_of_description(x):
    return len(x)

df['Length_of_Description'] = df['Description'].apply(get_length_of_description)
```

```python
def senior_or_not(x):
    if 'Senior' in x:
        return 1
    else:
        return 0

df['Senior'] = df['Job_Title'].apply(senior_or_not)
df
```

```python
df.groupby('Senior').count()
```

```python
y = df.groupby('Senior').mean()['Length_of_Description'].values
x = df.groupby('Senior').mean()['Length_of_Description'].index
```

```python
fig, ax = plt.subplots(figsize = (10,10))

ax.bar(x, y)
ax.set_xlabel('Senior Included in Job title or not')
ax.set_ylabel('Length of Description')
ax.set_xticks([0, 1])
ax.set_xticklabels(['Not Senior', 'Senior'])
ax.set_title('Length of Description for Jobs That Include Senior in Title or Not')

plt.tight_layout();
```

```python
df_all = pd.read_csv('../Datasets/df_all_linkedin.csv')
df_all['State']
```

```python
df_all[df_all['Job_Title'].str.contains('Junior')]
```

```python
companies = df_all['Company'].value_counts().index[:15]
companies = companies[::-1]

jobs = df_all['Company'].value_counts().values[:15]
jobs = jobs[::-1]

fig, ax = plt.subplots(figsize=(20,10))

ax.barh(companies, jobs)

ax.set_xlabel('Number of job postings')
ax.set_ylabel('Companies')
ax.set_title('Number of job postings for each company on Linkedin')
plt.tight_layout();
```

```python
df_all_co = df_all[df_all['State'] == 'CO']
companies_co = df_all_co['Company'].value_counts().index[:15]
companies_co = companies_co[::-1]

jobs_co = df_all_co['Company'].value_counts().values[:15]
jobs_co = jobs_co[::-1]

fig, ax = plt.subplots(figsize=(20,10))

ax.barh(companies_co, jobs_co)

ax.set_xlabel('Number of job postings')
ax.set_ylabel('Companies')
ax.set_title('Number of job postings for each company on Linkedin in Colorado')
plt.tight_layout();
```

```python
def graph_top15_companies(state_id, ax):
    df_all_state = df_all[df_all['State'] == state_id]
    companies_state = df_all_state['Company'].value_counts().index[:10]
    companies_state = companies_state[::-1]

    jobs_state = df_all_state['Company'].value_counts().values[:10]
    jobs_state = jobs_state[::-1]

#     fig, ax = plt.subplots()

    ax.barh(companies_state, jobs_state)

    ax.set_xlabel('Number of job postings')
    ax.set_ylabel('Companies')
    ax.set_title(f'Number of job postings for each company on Linkedin in {state_id}')
#     return ax
```

```python
list_of_locs = ['CO', 'CA', 'FL', 'NY', 'UT', 'WA']
plt.rcParams.update({'font.size': 70})

fig, axs = plt.subplots(3, 2, figsize=(100,100))

for loc, ax in zip(list_of_locs, axs.flatten()):
    graph_top15_companies(loc, ax)

plt.tight_layout()
```
